```python
dados = mineracao.groupby(["biome", "level_2"])[anos].sum()

# Aqui a intenção é separar a base em grupos usando as variáveis
# coletadas em mineracao.
# Utilizando o vetor anos criado anteriormente,
# o pandas soma os valores de cada grupo para cada ano.

plt.figure(figsize=(22,10))

# Para cada combinação biome + level_2
for (biome, level2), linha in dados.iterrows():

 plt.plot(anos,linha.values,marker="o",linewidth=2,label=f"{biome} - {level2}")

 plt.title("Evolução temporal por Bioma e Categorias",fontsize=16)

plt.xlabel("Ano", fontsize=12)
plt.ylabel("Valor", fontsize=12)

plt.xticks(rotation=45)

plt.grid(True)

plt.legend(bbox_to_anchor=(1.02, 1),loc="upper left",fontsize=12)

plt.tight_layout()

plt.show()
```

Alternando as cores do gráfico 

```python
dados = mineracao.groupby(["biome", "level_2"])[anos].sum()

# Paleta de cores 
cores = plt.cm.tab20.colors  # cm (Colormap), tab20 (Paleta de cores)
plt.figure(figsize=(24,12))
for i, ((biome, level2), linha) in enumerate(dados.iterrows()):

 plt.plot(anos,linha.values,color=cores[i % len(cores)],marker="o",linewidth=2,label=f"{biome} - {level2}")

plt.title("Evolução temporal por Bioma e Categorias",fontsize=18)
plt.xlabel("Ano", fontsize=14)
plt.ylabel("Valor", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(bbox_to_anchor=(1.02, 1),loc="upper left",fontsize=9)
plt.tight_layout()

plt.show()
```

 **Séries de Tempo** 

Dados em estrura de Série de Tempo representa um conjunto de dados que consiste em observações de uma ou várias variáveis **registradas sequencialmente ao longo do tempo** por uma determinada unidade.

*Exemplos: Preços de ações, oferta de moeda, Produto Interno Bruto (PIB), taxas de homicídio e vendas de automóveis*

**Para ilustrar o exemplo, vamos importar a série histórica das ações da Petrobras por meio da biblioteca yfinance (Yahoo Finance).**

Comando: pip install yfinance

```python
import yfinance as yf

# Baixar dados da ação
petr4 = yf.download("PETR4.SA",start="2015-01-01",end="2025-01-01")


print(petr4.head(40))
```

```python
print(petr4.columns)
```

Explicando as variáveis:

**Gráficos**

1) Gráfico do preço de fechamento

```python
plt.figure(figsize=(14,6))
plt.plot(petr4.index,petr4["Close"],linewidth=2)
plt.title("Preço da PETR4 ao longo do tempo")
plt.xlabel("Ano")
plt.ylabel("Preço de Fechamento")
plt.grid(True)
plt.show()
```

Para o proximo gráfico iremos criar a média movel 30 dias - (Calcula a Média dos ultimos 30 dias) 

```python
petr4["MM30"] = petr4["Close"].rolling(window=30,min_periods=1).mean() 
```

Gráfico com duas variáveis: MM30 e Close

```python

# Estrutura do Gráfico 
plt.figure(figsize=(15,6))

# Série original
plt.plot(petr4.index,petr4["Close"],label="Preço Diário",linewidth=1)

# Tendência suavizada
plt.plot(petr4.index,petr4["MM30"],label="Média Móvel 30 dias",linewidth=1)

plt.title("PETR4 - Série Temporal e Tendência")
plt.xlabel("Ano")
plt.ylabel("Preço")
plt.legend()
plt.grid(True)
plt.show()
```

**Dados em Painel (ou Dados Longitudinais)**

A principal característica que define os dados em painel é que as mesmas unidades (sejam indivíduos, empresas, domicílios ou cidades) **são acompanhadas e medidas ao longo de um determinado período de tempo, geralmente mês/ano**. 

Exemplo: Observar os indices de criminalidade das mesmas cidades,estados ou macroregiões durante um periodo de tempo (2000-01 até 2010-01). 

A vantagem de trabalhar com dados em painel é o **controle de características não observadas** ou seja, ter múltiplas observações da mesma unidade ajuda a controlar características intrínsecas que não podemos observar ou medir diretamente, facilitando muito a inferência de relações de causalidade (o que seria muito difícil usando apenas um cross section único).

*Para o primeiro exemplo, utilizaremos dados em painel para acompanhar o comportamento de diferentes instituições financeiras, segmentadas por estado e pertencentes ao mesmo setor, ao longo do tempo.*



**Bancos: JPMorgan (New York), Bank of America (North Carolina) e Wells fargo (California)**

```python
# Baixar as ações referente aos bancos:

# Vamos criar um vetor de identificação para cada banco para facilitar o download 

empresas = {
"JPMorgan": {"ticker": "JPM", "estado": "New York"},
"Bank of America": {"ticker": "BAC","estado": "North Carolina"},
"Wells Fargo": {"ticker": "WFC","estado": "California"},
"Citigroup": {"ticker": "C","estado": "New York"}}

# Lista para armazenar dados
lista = []

# Agora vamos criar um loop para baixar os dados referentes as ações de cada banco 

for empresa, info in empresas.items():
    
    print(f"Baixando: {empresa}") # Mostrar aviso que esta baixando

    dados = yf.download(
    info["ticker"],
    start="2015-01-01",
    end="2025-01-01",
    auto_adjust=True)

    #  Criar identificadores de colunas
    dados["empresa"] = empresa
    dados["estado"] = info["estado"] 

    # Transformar o Indice em Coluna (Lembrando o comando para entender qual é o indice usado na base de dados print(nomedabase.index) )
    dados = dados.reset_index()

    # Corrigir MultiIndex do yfinance
    dados.columns = [col[0] if isinstance(col, tuple) else col for col in dados.columns]

    # Adicionar à lista (Adicionar as empresas salvas na lista acima que criamos, nos dados)
    lista.append(dados)

# Construir o Painel 

painel = pd.concat(lista)

# Inserindo as variáveis no painel 

painel = painel[["empresa","estado","Date","Close","Volume"]]

# Mostrar a estrutura final 

print(painel.head(30))
print(painel["empresa"].unique())
```

**Transformar uma variável em Logaritmo**

```python
bwght['log_cigprice'] = np.log(bwght['cigprice'])

print(bwght["log_cigprice"].head(50))
```

**Quando é necessário transformar a variável e log Natural?**

- Assimetria nos Dados:
Motivação: Quando a distribuição da variável é assimétrica (não é simétrica), a transformação logarítmica pode ajudar a aproximá-la de uma distribuição mais simétrica.

- Redução de Amplitude:
Motivação: Se a variável abrange uma ampla gama de valores, a transformação logarítmica pode reduzir a amplitude dos dados, destacando variações em porcentagens ou proporções em vez de diferenças absolutas.

- Estabilidade de Variância:
Motivação: A transformação logarítmica pode estabilizar a variância, tornando-a mais constante ao longo da escala da variável. Isso é benéfico para modelos estatísticos que assumem homocedasticidade (variância constante).

- Interpretação Multiplicativa:
Motivação: Em alguns contextos, as diferenças percentuais são mais relevantes do que as diferenças absolutas. A transformação logarítmica converte as diferenças absolutas em diferenças percentuais, facilitando a interpretação multiplicativa.

- Linearização de Relacionamentos Não Lineares:
Motivação: Quando há um relacionamento exponencial ou logarítmico entre variáveis, a transformação logarítmica pode linearizar esses relacionamentos, tornando-os mais adequados para modelagem linear.

*Observações Importantes:*

- A transformação logarítmica é mais apropriada quando a variável contém apenas valores positivos.

- A interpretação dos resultados após a transformação deve levar em consideração a escala logarítmica.

- A escolha de transformar variáveis depende do contexto específico dos dados e dos objetivos da análise.
